```python
#pandas is a library and it contains the functions related to the database
import pandas as pd
```

```python
#Reading the "Salary_Data.csv" file into a pandas DataFrame named "dataset"
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#Creating a new DataFrame called "independent" containing only the "YearsExperience" column from the "dataset"
independent=dataset[["YearsExperience"]]
independent
```

```python
#Creating another new DataFrame called "dependent" containing only the "Salary" column from the "dataset"
dependent=dataset[["Salary"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
#Using scikit-learn's train_test_split function to split the data into training and testing sets.
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
#The data is split with 70% for training and 30% for testing, and random_state=0 is set to ensure reproducibility.
```

```python
#Creating a Linear Regression model using scikit-learn's LinearRegression class and fitting it to the training data.
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Calculating the coefficients (weights) of the trained Linear Regression model.
weight=regressor.coef_
weight
```

```python
#Calculating the intercept (bias) of the trained Linear Regression model.
bais=regressor.intercept_
bais
```

```python
#Using the trained Linear Regression model to make predictions on the test data.
y_pred=regressor.predict(X_test)
```

```python
#Calculating the R-squared score to evaluate the performance of the Linear Regression model.
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#Importing the pickle module and specifying a filename to save the trained Linear Regression model as a serialized object.
import pickle
filename="finalized_model_linear1.sav"
```

```python
#Serializing and saving the trained Linear Regression model using pickle.
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Loading a previously trained Linear Regression model and using it to make a prediction.
loaded_model=pickle.load(open("finalized_model_linear1.sav",'rb'))
result=loaded_model.predict([[13]])
```
